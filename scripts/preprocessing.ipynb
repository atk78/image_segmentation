```python
from pathlib import Path
import shutil
from sklearn.model_selection import train_test_split

import cv2

img_dir = Path("./../data/raw/img/")
label_dir = Path("./../data/raw/label/")
train_path = Path("./../data/train")
val_path = Path("./../data/valid")
test_path = Path("./../data/test")

img_file_list = sorted(list(img_dir.glob("*.bmp")))

# train_img_list = img_file_list[:200]

# val_img_list = img_file_list[200:224]
# test_img_list = img_file_list[224:]
train_img_list, test_img_list = train_test_split(
    img_file_list, test_size=0.2, random_state=42
)
val_img_list, test_img_list = train_test_split(
    test_img_list, test_size=0.5, random_state=42
)

if train_path.exists():
    shutil.rmtree(train_path)
if val_path.exists():
    shutil.rmtree(val_path)
if test_path.exists():
    shutil.rmtree(test_path)

train_path.mkdir(parents=True, exist_ok=True)
train_path.joinpath("img").mkdir(parents=True, exist_ok=True)
train_path.joinpath("mask").mkdir(parents=True, exist_ok=True)
val_path.mkdir(parents=True, exist_ok=True)
val_path.joinpath("img").mkdir(parents=True, exist_ok=True)
val_path.joinpath("mask").mkdir(parents=True, exist_ok=True)
test_path.mkdir(parents=True, exist_ok=True)
test_path.joinpath("img").mkdir(parents=True, exist_ok=True)
test_path.joinpath("mask").mkdir(parents=True, exist_ok=True)

for file in Path("./../data/raw/img").glob("*.bmp"):
    fname = file.stem
    img = cv2.imread(str(file))
    label = cv2.imread(str(label_dir.joinpath(fname + "_label.png")), cv2.IMREAD_GRAYSCALE)
    if file in train_img_list:
        cv2.imwrite(str(train_path.joinpath("img", fname + ".png")), img)
        cv2.imwrite(str(train_path.joinpath("mask", fname + ".png")), label)
    elif file in val_img_list:
        cv2.imwrite(str(val_path.joinpath("img", fname + ".png")), img)
        cv2.imwrite(str(val_path.joinpath("mask", fname + ".png")), label)
    elif file in test_img_list:
        cv2.imwrite(str(test_path.joinpath("img", fname + ".png")), img)
        cv2.imwrite(str(test_path.joinpath("mask", fname + ".png")), label)
```
