```python
from pathlib import Path
import shutil

import cv2

img_dir = Path("./../data/img/")
label_dir = Path("./../data/label/")
train_path = Path("./../data/train")
val_path = Path("./../data/valid")
test_path = Path("./../data/test")

img_file_list = sorted(list(img_dir.glob("*.bmp")))

train_img_list = img_file_list[:200]

val_img_list = img_file_list[200:224]
test_img_list = img_file_list[224:]

for file_path in train_img_list:
    shutil.copy(file_path, train_path.joinpath("img").joinpath(file_path.name))
    mask_path = label_dir.joinpath(f"{file_path.stem}_label.png")
    shutil.copy(mask_path, train_path.joinpath("label").joinpath(mask_path.name))

for file_path in val_img_list:
    shutil.copy(file_path, val_path.joinpath("img").joinpath(file_path.name))
    mask_path = label_dir.joinpath(f"{file_path.stem}_label.png")
    shutil.copy(mask_path, val_path.joinpath("label").joinpath(mask_path.name))

for file_path in test_img_list:
    shutil.copy(file_path, test_path.joinpath("img").joinpath(file_path.name))
    mask_path = label_dir.joinpath(f"{file_path.stem}_label.png")
    shutil.copy(mask_path, test_path.joinpath("label").joinpath(mask_path.name))
```

```python
def preconvert(label_train: Path, mask_train: Path):
    filelist = sorted(label_train.glob("*.png"))

    for file in filelist:
        mask = cv2.imread(str(file))
        mask[mask == 255] = 1  # 肺野  lung
        mask[mask == 85] = 2   # 心臓  heart
        mask[mask == 170] = 3   # 体   body

        fname = file.name.split("_")[0] + ".bmp"
        cv2.imwrite(mask_train.joinpath(fname), mask)


img_train_dir = Path("./../data/train/img/")
label_train_dir = Path("./../data/train/label/")
mask_train_dir = Path("./../data/train/mask/")
img_valid_dir = Path("./../data/valid/img/")
label_valid_dir = Path("./../data/valid/label/")
mask_valid_dir = Path("./../data/valid/mask/")
img_test_dir = Path("./../data/test/img/")
label_test_dir = Path("./../data/test/label/")
mask_test_dir = Path("./../data/test/mask/")

preconvert(label_train_dir, mask_train_dir)
preconvert(label_valid_dir, mask_valid_dir)
preconvert(label_test_dir, mask_test_dir)
```
